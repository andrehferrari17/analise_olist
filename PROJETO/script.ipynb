```python
# Importação dos pacotes

import pandas as pd
import duckdb
import os
```

```python
# Criação do banco de dados e importação dos arquivos .csv para dentro dele
con = duckdb.connect('olist.db')

caminho_pasta = '../CSV_FILES/'
arquivos = [f for f in os.listdir(caminho_pasta) if f.endswith('.csv')]

for arquivo in arquivos:
    nome_tabela = arquivo.replace('.csv', '')
    caminho_completo = os.path.join(caminho_pasta, arquivo)
    
    con.execute(f"CREATE TABLE {nome_tabela} AS SELECT * FROM read_csv_auto('{caminho_completo}')")
    print(f"Tabela {nome_tabela} criada com sucesso!")
```
